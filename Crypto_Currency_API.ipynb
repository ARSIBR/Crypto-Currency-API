```python
#Function to append data and save to a csv file
def api_runner():
    global df
    
    url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
    parameters = {
      'start':'1',
      'limit':'10',
      'convert':'INR'
    }
    headers = {
      'Accepts': 'application/json',
      'X-CMC_PRO_API_KEY': 'b1963c5e-a898-44f7-86c6-15b78b6296c8',
    }
    
    session = Session()
    session.headers.update(headers)
    
    try:
      response = session.get(url, params=parameters)
      data = json.loads(response.text)
      #print(data)
    except (ConnectionError, Timeout, TooManyRedirects) as e:
      print(e)
        
    df2 = pd.json_normalize(data['data'])
    df2['Timestamp'] = pd.to_datetime('now')
    df_append = pd.DataFrame(df2)
    df = pd.concat([df, df_append])

    if not os.path.isfile(r'C:\Users\HP\Downloads\CryptoCurrencyAPI.csv'):
        df.to_csv(r'C:\Users\HP\Downloads\CryptoCurrencyAPI.csv', header='column_names')
    else:
        df.to_csv(r'C:\Users\HP\Downloads\CryptoCurrencyAPI.csv', mode='a', header=False)
```

```python
import os
from time import time
from time import sleep

#Runs constantly every 60 seconds to collect data
while True:
    api_runner()
    print("Succesful API Run")
    sleep(60)
```

```python
#Tracking change over time
df3 = df.groupby('name', sort=False)[['quote.INR.percent_change_1h','quote.INR.percent_change_24h','quote.INR.percent_change_7d','quote.INR.percent_change_30d','quote.INR.percent_change_60d','quote.INR.percent_change_90d']].mean()
df3
```

```python
df4 = df3.stack()
df4
```

```python
df5 = df4.to_frame(name = 'values')
df5
```

```python
#set the index
df6 = df5.reset_index()
df6
```

```python
df7 = df6.rename(columns = {"level_1":"Change Period"})
df7['Change Period'] = df7['Change Period'].replace(['quote.INR.percent_change_24h','quote.INR.percent_change_7d','quote.INR.percent_change_30d','quote.INR.percent_change_60d','quote.INR.percent_change_90d'],['24h','7d','30d','60d','90d'])
df7
```

```python
sns.catplot(x='Change Period', y='values', hue='name', data=df7, kind='point', height=8, aspect=1.5)
```
